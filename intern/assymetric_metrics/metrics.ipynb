```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y_true = np.linspace(0, 1, 1000)
y_pred = np.linspace(0, 1, 1000)[::-1]

residuals = y_true - y_pred

def loss(residuals):
    return np.square(residuals)

loss_vals = loss(residuals)

plt.plot(residuals, loss_vals)
```

```python
y_true = np.linspace(0, 1, 1000)
y_pred = np.linspace(0, 1, 1000)[::-1]

residuals = y_true - y_pred

def loss(y_true, y_pred):
    # Like RELU but flipped horizontally
    return np.maximum((y_true - y_pred)*0.5, y_pred - y_true)

loss_vals = loss(y_true, y_pred)

plt.plot(residuals, loss_vals)
```
